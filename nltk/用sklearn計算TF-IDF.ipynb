```python
from sklearn.feature_extraction.text import TfidfVectorizer
import jieba
import pandas as pd

#chinese data list
data = ["其實我很愛你，你知道嗎？", "給你一個讚阿", "我愛你到不要不要的"]

#take a look at segmented chinese texts
print( " , " .join(jieba.cut(data[0])))

def jieba_tokenizer(x):
    a= " " .join(jieba.cut(x))
    return a

# tf idf with ngram_range 1 to 3 
vect = TfidfVectorizer(tokenizer=jieba_tokenizer,ngram_range=(1,3)) 

# value of tf-idf of each ngrammed text
X_train_features = vect.fit_transform(data)
idf = vect.idf_
print(dict(zip(vect.get_feature_names(), idf)))

#make pd data frame
gg=pd.DataFrame(list(X_train_features)) #list df
gg=pd.DataFrame(list(X_train_features.toarray())) #sparse df
```
